```python
import matplotlib.pyplot as plt

plt.plot([2,4,6,4])
plt.ylabel("numbers")
plt.xlabel("indices")
plt.title("My plot")
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25])
plt.ylabel("squares")
plt.xlabel("numbers")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4,5], [1,4,9,16,25],'ro')
plt.ylabel("squares")
plt.xlabel("numbers")
plt.grid()
plt.show()
```

```python
import numpy as np

x= np.arange(1,5,1)

plt.plot(x, x**2, 'bs', label='^2' )
plt.plot(x, x**3, 'g^', label='^3')
plt.plot(x, x**4, 'r--', label='^4')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y, linewidth=10)
plt.show()
```

```python
x1=[1,2,3,4,5]
y1=[1,4,9,16,25]
x2=[1,2,3,4,5]
y2=[2,4,6,8,10]

multiple_lines=plt.plot(x1,y1,x2,y2)
plt.setp(multiple_lines[0], color='g', linewidth=5)
plt.setp(multiple_lines[1], color='b', linewidth=8)
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'g--')


plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r-')
plt.show()
```
